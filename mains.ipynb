```python
from tabula import read_pdf
import pandas as pd




tabelas = read_pdf("Sistemas de Informação.pdf", pages='all')
print(len(tabelas))
```

```python

#print(tabelas.columns)
```

```python

tabela = tabelas[0].copy()

tabela.columns = tabela.iloc[0]

tabela = tabela[0:11].copy()  

tabela.columns = tabela.iloc[0]
tabela = tabela[1: ]


# Lista com os nomes desejados para as três primeiras colunas
nomes_colunas = ['Ano-Período', 'Ingressantes', 'Ativos']

# Atribuindo nomes às três primeiras colunas
tabela.columns = nomes_colunas + list(tabela.columns[3:])


tabela = tabela.set_index("Ano-Período")

# Convertendo colunas 'Ingressantes' e 'Ativos' para inteiros
#tabela['Ingressantes'] = tabela['Ingressantes'].astype(int)
#tabela['Ativos'] = tabela['Ativos'].astype(int)

pd.set_option('display.max_columns', None)

tabela1 = tabela

tabela.to_csv('Sistemas de Informação-Trancamentos.csv')

display(tabela1)
```
